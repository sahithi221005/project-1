```python
import pandas as pd
```

```python
# Read the Excel file into a pandas DataFrame
# Replace 'path_to_file.xlsx' with the actual path to your file
file_path = '/content/drive/MyDrive/question_1.xlsx'
df = pd.read_excel(file_path, engine="openpyxl")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Create new columns for Total, Maximum, and Average marks
df['Total'] = df[['math score', 'reading score', 'writing score']].sum(axis=1)
```

```python
df['Maximum'] = df[['math score', 'reading score', 'writing score']].max(axis=1)
```

```python
df['Minimum'] = df[['math score', 'reading score', 'writing score']].min(axis=1)
```

```python
df['Average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```

```python
import matplotlib.pyplot as plt

df['Total'].plot(kind='hist', bins=10, edgecolor='blue', title='Histogram of Total Marks')

plt.xlabel('Total Marks')
plt.ylabel('frequency')

plt.show()
```
